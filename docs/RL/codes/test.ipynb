```python
import numpy as np
```

```python
rewards = np.arange(0, 1, 0.05, dtype=np.float64)
```

```python
G = len(rewards)
```

```python
mean = np.mean(rewards)
std = np.std(rewards)
```

```python
advantages = (rewards - mean)/std
```

```python
r_theta = np.random.rand(20) + 1
```

```python
advantages, r_theta
```

```python
EPS = 0.1

t1 = r_theta * advantages
t2 = np.clip(r_theta, 1 - EPS, 1 + EPS) * advantages
```

```python
L_clipped = np.min([t1, t2], axis=0)
```
